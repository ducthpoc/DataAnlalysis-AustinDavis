# **Two types of data:**

1.   *Numerical*
2.   *Categorical*






# **Numercical Data**


1.   Discrete - Can only be certain fixed values
2.   Continuous - Can be any real number (infinite possible values)

# Examples of discrete data:


*   Counting students in a classroom, can only have whole numbers (1, 2, 3, etc), decimals don't work.
*   Looking at shoe sizes, only move up in increments of 0.5, so a shoe size of 7.364729 wouldn't exist.

# Examples of continuous data:


*   Measuring the height of students, height could be any number, from 6 foot even, to 5.139684 feet, to 5.8964 feet, etc.
*   Distance traveled in a race, could be measured infinitely precisely by mm, cm, m, km, etc.
