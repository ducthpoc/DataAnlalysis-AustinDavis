```python
# Importing libraries we're going to use
import pandas as pd
from google.colab import files
import io

# Upload file and turn it into a DataFrame
dataFile = files.upload()
dataset = pd.read_csv(io.BytesIO(dataFile['AI_index_db.csv']))
```

```python
# View first dataset entries
dataset.head()
```

```python
# Describe function
dataset.describe()
```

```python
# Singular column selection
dataset["Research"]
```

```python
# Multiple column selection
talentDataframe = dataset[["Country", "Talent"]]
talentDataframe
```

```python
# loc - Location
# iloc - Integer Location
```

```python
# Selecting specific cell/row
dataset.loc[0]
```

```python
# Selecting multiple rows/columns
dataset.loc[:, "Research":]
```

```python
# Selecting rows/columns based on a condition (Filtering)
dataset.loc[dataset["Region"].isin(["Europe", "Americas", "Middle East"])]
```

```python
# Selecting rows/columns based on MULTIPLE conditions
dataset.loc[(dataset["Region"].isin(["Europe", "Americas", "Middle East"])) & (dataset["Operating Environment"] < 70)]
```

```python
# Multiple Filtering Example 2
dataset.loc[(dataset["Infrastructure"] > 50) & (dataset["Region"] == "Asia-Pacific")]
```
