```python
import pandas as pd
from google.colab import files
import io
import matplotlib.pyplot as plt

dataFile = files.upload()
print(dataFile)

dataset = pd.read_csv(io.BytesIO(dataFile['AI_index_db.csv']))
```

```python
dataset.head()
```

```python
dataset.mean()
```

```python
# Mean - Average

# # Example: 50, 60, 70, 80, 90
# # 50 + 60 + 70 + 80 + 90 = 350
# # 350 / 5 = 70
# # Mean = 70
```

```python
# Median - Middle number when sorted

# # Example: 40, 10, 28, 7, 85
# # Sorted: 7, 10, 40, 85
# # Median = 25
```

```python
# Variance - Squared difference of each data point measured by mean

# # Example: 2, 3, 7, 9, 12
# # Mean = (2 + 3 + 7 + 9 + 12) / 5 = 6.6
# # (2 - 6.6)^2 = 21.16
# # (21.16 + 12.96 + 0.16 + 5.76 + 29.16) / 5 = 13.84
# # Variance = 13.84
```

```python
# Standard deviation - Square root of variance

# # Example - Above problem
# # Variance = 13.84
# # Standard Deviation = sqrt(13.84)
# # Standard Deviation = 3.72
```

```python
# Finding Mean
dataset["Talent"].mean()
```

```python
# Finding Median
dataset.loc[:, "Talent":"Development"].median()
```

```python
# Finding Variance
dataset[["Talent", "Research", "Total score"]].var()
```

```python
# Finding Standard Deviation
dataset["Government Strategy"].std()
```

```python
# Correlation Plotting
import seaborn as sns

sns.pairplot(dataset)
```

```python
plt.scatter(dataset["Research"], dataset["Total score"])

plt.xlabel("Research")
plt.ylabel("Total score")
plt.title("Total Score Vs Research")

plt.show()
```
