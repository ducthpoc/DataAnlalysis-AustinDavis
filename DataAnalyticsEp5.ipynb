```python
import pandas as pd
from google.colab import files
import io
import matplotlib.pyplot as plt

dataFile = files.upload()
dataset = pd.read_csv(io.BytesIO(dataFile['AI_index_db.csv']))
```

```python
dataset.head()
```

```python
# Basic line plot with model data
x = list(range(0,10))
y = list(range(10, 20))

plt.plot(x, y, color = "red")

plt.xlabel("Insert x label here")
plt.ylabel("Insert y label here")

plt.show()
```

```python
# Basic scatterplot with model data
x = [1, 3, 6, 8, 14]
y = [4, 7, 7, 10 ,12]

plt.scatter(x, y, color = "red")

plt.xlabel("Insert x label here")
plt.ylabel("Insert y label here")

plt.show()
```

```python
# Basic bar graph with model data
labels = ["A", "B", "C", "D"]
values = [10, 20, 30, 40]
colors = ["red", "green", "blue", "purple"]

plt.barh(labels, values, color = colors)
plt.ylabel("Label Axis")
plt.xlabel("Value Axis")
plt.title("Graph Title")

plt.show()
```

```python
# Basic histogram with model data
import numpy as np

dataPoints = np.random.normal(100, 10, 100)

plt.hist(dataPoints)

plt.show()
```

```python
# Line graph with AI data
xvar = "Research"
x = dataset[xvar]
y = dataset["Total score"]

plt.plot(x, y)

plt.ylabel("Total Score")
plt.xlabel(xvar)
plt.title("Total Score vs " + xvar)

plt.show()
```

```python
# Scatterplot with AI data
xvar = "Talent"
x = dataset[xvar]
y = dataset["Total score"]

plt.scatter(x, y)

plt.ylabel("Total Score")
plt.xlabel(xvar)
plt.title("Total Score vs " + xvar)

plt.show()
```

```python
# Bar graph with AI data
incomeCounts = dataset.value_counts("Income group")

labels = list(incomeCounts.index)
values = list(incomeCounts)
colors = ["green", "red", "blue"]

plt.barh(labels, values, color = colors)
plt.ylabel("Income Group")
plt.xlabel("Count")
plt.title("Counts of Each Income Group")

plt.show()
```

```python
# Histogram with AI Data
values = dataset["Research"]

plt.hist(values)
plt.xlabel("Total Score")
plt.ylabel("Frequency")
plt.title("Frequency of All Total Scores")

plt.show()
```
